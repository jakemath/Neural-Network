```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('cost.txt', header=None, skiprows=lambda x: x % 1000000 != 0)
df = df[0].astype(str).str.split(' ', expand=True).astype(float)
df.columns = ['iter', 'cost']
df['iter'] = df['iter'].astype(int)
plt.plot(df['iter'], df['cost'])
```
